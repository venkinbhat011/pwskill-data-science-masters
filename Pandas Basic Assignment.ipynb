## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

### DateFrame
#### A DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).
#### It can be thought of as a collection of Pandas Series where each Series represents a column.
#### DataFrame is suitable for handling more complex, structured data, and it can store data of different data types.
#### It provides methods for various operations on the entire dataset or specific columns.

### Series:
#### A Series is a one-dimensional labeled array capable of holding any data type.
#### It represents a single column or row of data in a DataFrame.
#### Series is useful for handling simpler, one-dimensional data structures.
#### It is essentially a single column or row with an index.

## Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd
```

```python
# Creating multiple Series
names_series = pd.Series(['Akshata', 'Bharat', 'chandini'], name='Name')
ages_series = pd.Series([25, 30, 27], name='Age')
genders_series = pd.Series(['Female', 'Male', 'Female'], name='Gender')
```

```python
# Creating a DataFrame using the Series
df = pd.DataFrame({
    'Name': names_series,
    'Age': ages_series,
    'Gender': genders_series
})
```

```python
# Print the DataFrame
print(df)
```
